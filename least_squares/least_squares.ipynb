直线 $$ y=wx+b $$

$$L=\sum_{i=1}^{n}(wx_{i}+b-y_{i})^2$$

求使得L最小的w，b

![距离平方和最小](./d.png)

```python
# 先写一个计算loss的函数

def computer_L(w,b,x,y):
    loss_total = 0
    for i in range(len(x)):
        loss_total += (w*x[i]+b-y[i])**2
    return loss_total

computer_L(2,3,x,y) # 目的是找出w，b使得L最小，使得computer_L(w,b,x,y) 最小
```

```python
# 先写一个平均数计算函数

def average(data):
    count =  0
    for i in range(len(data)):
        count  += data[i]
    return count/len(data)

average([1,2,3])
```

```python
def fix(x,y):
    up = 0 #分子
    down = 0 # 分母
    x_bar = average(x)
    y_bar = average(y)
    
    
    for i in range(len(x)):
        up += (x[i]-x_bar)*(y[i]-y_bar)
        down += (x[i]-x_bar)**2
    
    w = up/down
    
    
    # 接着计算b
    
    b = y_bar - w * x_bar
    
    return w,b

w,b = fix(x,y)
print("w is",w) # 应该是3左右
print("b is",b) # 应该是10左右
    
    
```

```python
plt.scatter(x,y,c='c')
y_pre = w * x + b
plt.plot(x,y_pre,c='r')
plt.show()
```
